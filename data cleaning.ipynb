```python
import csv
import pymysql

# Establish the connection
myconnection = pymysql.connect(host='localhost', user='root', passwd='Vinodha@123', database="redbus")
cursor = myconnection.cursor()

# Create table (if not exists to avoid errors if it already exists)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS dbusdata(
        Id INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
        Route_Name VARCHAR(255),
        Route_Link VARCHAR(255),
        Bus_Name VARCHAR(255),
        Bus_Type VARCHAR(255),
        Departing_Time VARCHAR(20),
        Duration VARCHAR(255),
        Reaching_Time VARCHAR(20),
        Star_Rating FLOAT,
        Price FLOAT,
        Seat_Availability INT
    )
''')

# Open the CSV file
with open('C:/Users/ncssa/Downloads/bus_routes.csv', newline='') as csvfile:
    csv_data = csv.reader(csvfile)
    
    # Insert rows from the CSV file into the table
    for row in csv_data:
        if row!=['Route_Name', 'Route_Link', 'Bus_Name', 'Bus_Type', 'Departing_Time', 'Duration', 'Reaching_Time', 'Star_Rating', 'Price', 'Seat_Availability']:
            
            cursor.execute('''INSERT INTO dbusdata (Route_Name, Route_Link, Bus_Name, Bus_Type, Departing_Time,
                Duration, Reaching_Time, Star_Rating, Price, Seat_Availability)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        ''', row)

# Commit the transaction
myconnection.commit()

# Close the cursor and connection
cursor.close()
myconnection.close()

print("Done")
```
